```python
import numpy as np
import tensorflow.compat.v1 as tf

tf.disable_v2_behavior()

# XOR dataset
inputs = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
targets = np.array([[0], [1], [1], [0]], dtype=np.float32)

# Placeholders
x_ph = tf.placeholder(tf.float32, shape=[None, 2])
y_ph = tf.placeholder(tf.float32, shape=[None, 1])

# Weights and biases
w_hidden = tf.Variable(tf.random_normal([2, 2], stddev=0.1))
b_hidden = tf.Variable(tf.zeros([2]))
w_out = tf.Variable(tf.random_normal([2, 1], stddev=0.1))
b_out = tf.Variable(tf.zeros([1]))

# Forward pass
layer1 = tf.sigmoid(tf.matmul(x_ph, w_hidden) + b_hidden)
pred = tf.sigmoid(tf.matmul(layer1, w_out) + b_out)

# Loss + optimizer
mse = tf.reduce_mean(tf.square(pred - y_ph))
train_op = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(mse)

# Training
with tf.Session() as session:
    session.run(tf.global_variables_initializer())

    for i in range(10000):
        session.run(train_op, feed_dict={x_ph: inputs, y_ph: targets})

        if i % 2000 == 0:
            current_loss = session.run(mse, feed_dict={x_ph: inputs, y_ph: targets})
            print("Step:", i, "Loss:", current_loss)

    print("Predicted Output:")
    print(session.run(pred, feed_dict={x_ph: inputs}))
```
